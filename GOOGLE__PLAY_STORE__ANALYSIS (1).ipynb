```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('/content/googleplaystore.csv')
```

```python


# Convert columns 'Reviews' and 'Installs' to integers
df[['Reviews', 'Installs']] = df[['Reviews', 'Installs']].replace({'[+,]': ''}, regex=True).apply(pd.to_numeric, errors='coerce').astype('Int64')

```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Category', data=df, palette='viridis')

# Adding labels and title
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Number of Apps by Category', fontsize=14)

# Rotating x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
avg_installs = df.groupby('Category')['Installs'].mean().sort_values(ascending=False)

# Plotting a bar chart
plt.figure(figsize=(12, 8))
sns.barplot(x=avg_installs.values, y=avg_installs.index, palette='viridis')

# Adding labels and title
plt.xlabel('Average Installs', fontsize=12)
plt.ylabel('Category', fontsize=12)
plt.title('Average Installs by App Category', fontsize=14)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
from scipy import stats


rating_free = df[df['Type'] == 'Free']['Rating']
rating_paid = df[df['Type'] == 'Paid']['Rating']

# Check assumptions: normality and equal variance (not shown in code)
# Conduct two-sample t-test assuming equal variances
t_statistic, p_value = stats.ttest_ind(rating_free, rating_paid, equal_var=True)

# Print the results
print(f"Two-sample t-test results:")
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in ratings between free and paid apps.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in ratings between free and paid apps.")
```

```python

# Convert 'Last Updated' to datetime format
df['Last Updated'] = pd.to_datetime(df['Last Updated'])

# Group by 'Last Updated' and calculate average rating
avg_rating = df.groupby(pd.Grouper(key='Last Updated', freq='M'))['Rating'].mean().reset_index()

# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Last Updated', y='Rating', data=avg_rating, marker='o', color='b')

# Adding labels and title
plt.xlabel('Last Updated', fontsize=12)
plt.ylabel('Average Rating', fontsize=12)
plt.title('Average Rating Over Time', fontsize=14)

# Display the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
```
