```python

def biplot(df, x, title, hue):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams["axes.labelsize"] = 20
    plt.rcParams['axes.titlesize'] = 22
    plt.rcParams['axes.titlepad'] = 30

    fig, ax = plt.subplots(figsize=(12, 8))
    sns.countplot(data=df, x=x, hue=hue, palette='bright', ax=ax)

    plt.title(title)
    plt.xlabel(x)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.legend(title=hue)

    plt.show()

# Example usage:
# Assuming new_df1_target1 is your dataframe and 'MultipleLines' and 'InternetService' are categorical variables, and 'gender' is also a categorical variable.
biplot(new, x='MultipleLines', title='Churn Status by Multiple Lines and Internet Service', hue='InternetService')
```

```python
# Convert relevant columns to numeric (handle errors if necessary)
df['MonthlyCharges'] = pd.to_numeric(df['MonthlyCharges'], errors='coerce')
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
df['tenure'] = pd.to_numeric(df['tenure'], errors='coerce')

# Select numeric variables and calculate correlation
numerical_vars = ['MonthlyCharges', 'TotalCharges', 'tenure']
numerical_corr = df[numerical_vars].corr()

# Plotting the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(numerical_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix (Numerical Variables)', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
for var in numerical_vars:
    sns.histplot(df, x=var, hue='Churn', bins=20, kde=True, palette='Set2', alpha=0.7)
plt.title('Distribution of Numerical Variables by Churn', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(title='Churn', fontsize=10, title_fontsize='12')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder

# Drop non-numeric columns and identifiers
X = df.drop(['Churn'], axis=1)  # Assuming CustomerID is non-numeric
y = df['Churn']

# Handle missing values (NaNs) in numeric columns using SimpleImputer with mean strategy
numeric_cols = X.select_dtypes(include='number').columns
imputer = SimpleImputer(strategy='mean')
X[numeric_cols] = imputer.fit_transform(X[numeric_cols])

# Convert categorical variables to numeric using LabelEncoder if needed
le = LabelEncoder()
for col in X.select_dtypes(include='object'):
    X[col] = le.fit_transform(X[col])

# Initialize and fit the RandomForestClassifier
clf = RandomForestClassifier(random_state=42)
clf.fit(X, y)

# Extract feature importances
feature_importances = pd.Series(clf.feature_importances_, index=X.columns)

# Plot the top 10 feature importances
plt.figure(figsize=(10, 6))
feature_importances.nlargest(10).plot(kind='barh', figsize=(10, 6))
plt.title('Top 10 Feature Importances for Churn Prediction', fontsize=14)
plt.xlabel('Feature Importance', fontsize=12)
plt.ylabel('Feature', fontsize=12)
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer

# Step 1: Handle missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(df[numerical_vars])

# Step 2: Scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Step 3: Apply KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)

# Step 4: Visualize the clusters
sns.countplot(data=df, x='cluster', hue='Churn', palette='Set2')
plt.title('Churn Distribution by Customer Clusters', fontsize=14)
plt.xlabel('Cluster', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(fontsize=10)
plt.legend(title='Churn', fontsize=10, title_fontsize='12')
plt.show()
```
